```python
# tableau des mois en français
french_months =  ["Janvier", "Février", "Mars   ", "Avril", "Mai  ", "Juin ", "Juillet  ", "Août     ", "Septembre", "Octobre ", "Novembre", "Décembre"]

# tableau des mois en anglais
english_months = ["January ", "February", "March   ", "April", "May  ", "June ", "July     ", "August   ", "September", "October ", "November", "December"]

# Affichage des mois en français par défaut
print("Mois en français :")
for i in range(0, 12 ,3):
    print("\t",french_months[i] , end=" | ")
print("\n")
for i in range(1, 12 ,3):
    print("\t",french_months[i] , end=" | ")
print("\n")
for i in range(2, 12 ,3):
    print("\t",french_months[i] , end=" | ")
print("\n")
# boucle de menu

while True:
    print("\n Menu :")
    print("1. Mois en Français")
    print("2. Mois en Anglais")
    print("3. Quitter")
    
    # demande de choix de l'utilisateur
    choice = int(input("Entrez votre choix (1, 2 ou 3) : "))
    
    # action en fonction du choix de l'utilisateur
    if choice == 1:
        print("\nMois en français :")
        for i in range(0, 12 ,3):
           print("\t",french_months[i] , end=" | ")
        print("\n")
        for i in range(1, 12 ,3):
           print("\t",french_months[i] , end=" | ")
        print("\n")
        for i in range(2, 12 ,3):
           print("\t",french_months[i] , end=" | ")
        print("\n")
    elif choice == 2:
        print("\nMois en anglais :")
        for i in range(0, 12 ,3):
            print("\t",english_months[i] , end=" | ")
        print("\n")
        for i in range(1, 12 ,3):
            print("\t",english_months[i] , end=" | ")
        print("\n")
        for i in range(2, 12 ,3):
            print("\t",english_months[i] , end=" | ")
        print("\n")
    elif choice == 3:
        print("Au revoir!")
        break
    else:
        print("Entrée non valide. Veuillez entrer 1, 2 ou 3.") 
```
